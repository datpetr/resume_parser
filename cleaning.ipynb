import necessary libraries 

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

download the necessary NLTK data files

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

Convert text to lowercase, remove punctuation, special characters, and stopwords

```python
import re
```

```python
def normalize_text(text):
    # Remove years (4 consecutive digits)
    text_without_years = re.sub(r'\b\d{4}\b', '', text)
    # Remove special characters except for spaces
    pattern = r'[^a-zA-Z0-9\s]'
    clean_text = re.sub(pattern, '', text_without_years)
    # Replace any kind of line breaks with a space
    cleaned_text = re.sub(r'[\r\n]+', ' ', clean_text)
    # Remove extra spaces
    final_text = re.sub(r'\s+', ' ', cleaned_text).strip()
    return final_text
```

Tokenize text into words and apply lemmatization

```python
def tokenize_and_lemmatize(text):
    # Tokenize text into words
    tokens = word_tokenize(text)
    
    # Initialize lemmatizer
    lemmatizer = WordNetLemmatizer()
    
    # Apply lemmatization
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    return lemmatized_tokens
```

apply text normalization and tokenization

```python
df['Normalized_Resume'] = df['Resume'].apply(normalize_text)
df['Normalized_Resume']
```

```python
df['Tokenized_Lemmatized_Resume'] = df['Normalized_Resume'].apply(tokenize_and_lemmatize)
df['Tokenized_Lemmatized_Resume']
```

## Feature extraction
### extract key features
Identify and extract such as skills, jib titles, and relevant terms 

```python
from nltk import pos_tag, ne_chunk
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def extract_features(text):
    # Tokenize and POS tag the text
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    
    # Extract skills and job titles (assuming they are proper nouns)
    skills = [word for word, pos in pos_tags if pos == 'NNP']
    
    return skills
```

Apply the feature extraction function to the Normalized_Resume column

```python
df['Extracted_Features'] = df['Normalized_Resume'].apply(extract_features)
df['Extracted_Features']
```

## Clustering and Skill Grouping
First, we will choose a clustering algorithm. Let's use K-means clustering for this example. We'll also prepare the necessary libraries and preprocess the data using TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Limit features to avoid overfitting

# Fit and transform the normalized resume text
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Normalized_Resume'])

tfidf_matrix.shape
```

Now we will implement and train the K-means clustering model

```python
# Choose the number of clusters (this can be tuned based on silhouette score or domain knowledge)
num_clusters = 10

# Initialize the K-means model
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Train the model on the TF-IDF matrix
kmeans.fit(tfidf_matrix)

# Get the cluster labels
cluster_labels = kmeans.labels_

# Add the cluster labels to the dataframe
df['Cluster_Labels'] = cluster_labels

# Display the first few rows with cluster labels
df.head()
```

We will evaluate the clustering results using the silhouette score and analyze the clusters to ensure meaningful grouping of skills and terms.

```python
silhouette_avg = silhouette_score(tfidf_matrix, cluster_labels)
```

Function to print the top terms per cluster

```python
def print_top_terms_per_cluster(tfidf_matrix, cluster_labels, num_clusters, terms):
    order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
    
    for i in range(num_clusters):
        print("\nCluster %d:" % i)
        for ind in order_centroids[i, :10]:  # Print the top 10 terms per cluster
            print(' %s' % terms[ind])
```

```python
# Get the terms from the TF-IDF vectorizer
terms = tfidf_vectorizer.get_feature_names_out()

# Print the top terms per cluster
print_top_terms_per_cluster(tfidf_matrix, cluster_labels, num_clusters, terms)
```
